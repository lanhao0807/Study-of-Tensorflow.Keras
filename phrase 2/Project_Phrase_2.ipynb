```python
import pandas as pd
data = pd.read_csv('Maternal_Health_Risk_Data_Set.csv', delimiter = ',')
data.head()
# Binary classification is done in phrase 1
```

```python
import numpy as np
dataset = np.genfromtxt('Maternal_Health_Risk_Data_Set.csv',delimiter=',', skip_header = True)
np.set_printoptions(formatter = {'float': '{:0.2f}'.format})
print(dataset.shape)
```

```python
print(dataset[:5, :]) # with fixed decimal place and skipped header
```

```python
X=dataset[:, :-1] 
Y=dataset[:,-1]
# separate input and output columns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(1, input_dim=len(X[0,:]), activation='sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics='accuracy' )
```

```python
model.fit(x= X, y=Y, epochs = 256, verbose = 1)
```

```python
print(Y[:10])
```

```python
predictions = model.predict(X)
print(predictions[:10].T)
```

```python
# as we can see here, this model makes mistakes and the single layer model's
# accuracy is not increasing no matter how many epochs repeated. So, now let's
# do it within multiple layers
```

```python
mean = X.mean(axis = 0)
X -= mean
range = X.max(axis = 0) - X.min(axis = 0)
X /= range
# mean normalization just like in phrase 1
```

```python
model2= Sequential()
model2.add(Dense(16, input_dim= len(X[0,:]), activation = 'relu' ))
model2.add(Dense(8, activation = 'relu' ))
model2.add(Dense(4, activation = 'relu' ))
model2.add(Dense(1, activation = 'sigmoid')) 
```

```python
model2.summary()
```

```python
model2.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics='accuracy' )
# build for multi-layered model
```

```python
model2.fit(x= X, y=Y, epochs = 2560, verbose = 1) # 4 times
```

```python
print(Y[:10])
```

```python
predictions2 = model2.predict(X)
print(predictions2[:10].T)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
my_accuracy= accuracy_score(Y, predictions2.round())
```

```python
print(my_accuracy)
```
